```python
#!/usr/bin/env python3
# In[48]:

import random
import operator

class genetic:
    def __init__(self):
        self.max_weight = 21
        self.size = 300 # tamanho da população
        self.size_itens = 6 # quantidade de itens na mochila
        self.pop = []
        for i in range(self.size):
            self.pop.append(self.create_population())
        self.fitness(self.pop)
        for i in range(200):
            self.crossover()
            self.elimina_pop()
        # self.print_(self.pop) # imprime todos os resultados
        print(self.pop[0])

    def elimina_pop(self):
        self.pop = self.pop[:self.size]
    def fitness(self, pop):
        max_value = 0
        max_weight = 0
        for i in range(len(pop)):
            for j in range(len(pop[i])):
                max_value += pop[i][j][1]
                max_weight += pop[i][j][2]
            if max_weight > self.max_weight:
                max_weight = 0
            pop[i].append(max_weight)
            pop[i].append(max_value)
            max_weight = 0
            max_value = 0
        pop.sort(key=operator.itemgetter(-1), reverse = True)

        for i in range(len(pop)):
            # print(pop[i][-2])
            if (pop[i][-2]) == 0:
                x = pop[i]
                del pop[i]
                pop += [x]
    def cria_filho(self, p1, p2, ordem = 1):
        if ordem == 1:
            half = self.size_itens//2
            filhos = []
            time = 0
            # x = random.randint(1,4)
            filhos = p1[:half]
            while (len(filhos) < self.size_itens) and (time < 100):
                r = random.randint(0, self.size_itens-2)
                if p2[r] not in filhos:
                    filhos.append(p2[r])
                time += 1
        return filhos
    def crossover(self):
        for i in range(0,len(self.pop),2):
            self.pop[i] = self.pop[i][:-2]
            self.pop[i+1] = self.pop[i+1][:-2]
            a = self.cria_filho(self.pop[i], self.pop[i+1])
            b = self.cria_filho(self.pop[i+1], self.pop[i])
            self.pop += [a]
            self.pop += [b]
        self.fitness(self.pop)
    def print_(self, pop):
        for i in range(len(pop)):
                print(pop[i])
    def create_population(self, max = 0):
        max = self.size_itens
        pop = []
        l = []
        for i in range(max):
            x = random.randint(1,12)
            while(True):
                if x not in l:
                    l.append(x)
                    break
                else:
                    x = random.randint(1,12)
            if x == 1:
                pop.append(['bug repelent', 12, 2])
            elif x == 2:
                pop.append(['camp stove', 5, 4])
            elif x == 3:
                pop.append(['canteen', 10, 7])
            elif x == 4:
                pop.append(['cothes', 11, 5])
            elif x == 5:
                pop.append(['dried food', 50, 3])
            elif x == 6:
                pop.append(['first-aid kit', 15, 3])
            elif x == 7:
                pop.append(['flashlight', 6, 2])
            elif x == 8:
                pop.append(['novel', 4, 2])
            elif x == 9:
                pop.append(['rain gear', 5, 2])
            elif x == 10:
                pop.append(['sleeping bag', 25, 3])
            elif x == 11:
                pop.append(['tent', 20, 11])
            elif x == 12:
                pop.append(['water filter', 30, 1])
        return pop



g = genetic()


# In[ ]:
```
